<h1>Doubly Linked List</h1>
<hr/>

<h2>Definition</h2>
<ol>
<li>collection of **nodes** that form a linear sequence</li>
<li>each node stores a **reference** to the **object** that it represents, as well as **references** to the **next node AND previous node**</li>
<li>also contains **header(first)** and **tailer(last)** nodes, which are **dummies nodes that don't contain objects** (in singly linked list head/tail are pointers)</li>
<li>**tailer node's next** and **header node's prev** must point to **None**</li>
<li>header and trailer nodes are known as **sentinels** aka **guards**</li>
</ol>

<h2>Operations</h2>

<ul>
<li>**Inserting** is easy:
<ol>
<li>create a new node</li>
<li>set its element to the new element</li>
<li>set its next link to next node</li>
<li>set its prev link to prev node</li>
<li>set prev nodes' next to new node</li>
<li>set next nodes' prev to new node</li>
</ol>
</li>
<li>**Deleting** is also easy
<ol>
<li>find node</li>
<li>set next nodes' prev to prev node</li>
<li>set prev nodes' next to next node</li>
<li>delete node</li>
</ol>
</li>
<li>
However **deleting last element** is hard because you do not readily have access to the previous element (you need a **doubly linked list** for this
</li>
</ul>
